# **Introdução**

O setor de saúde suplementar desempenha um papel fundamental na assistência à população brasileira, complementando o Sistema Único de Saúde (SUS). Segundo dados da Agência Nacional de Saúde Suplementar (ANS), mais de 50 milhões de brasileiros possuem algum tipo de plano de saúde, representando aproximadamente 24% da população. Esse número apresentou crescimento expressivo nos últimos anos, impulsionado especialmente pela pandemia de Covid-19, que aumentou a demanda por acesso rápido e especializado aos serviços de saúde.

Apesar desse crescimento, observa-se uma distribuição desigual dos planos de saúde no país, com variações significativas de acordo com fatores como faixa etária, localização geográfica e tipo de contratação. Regiões do Sudeste concentram aproximadamente 65% dos beneficiários, enquanto Norte e Nordeste apresentam taxas de adesão consideravelmente menores. Ademais, fatores econômicos e variações nos custos impactam diretamente a expansão e a sustentabilidade do setor. Diante desse cenário, surge a necessidade de compreender quais fatores impulsionam ou limitam o crescimento do setor de planos de saúde no Brasil.

A motivação para este estudo decorre da percepção de desigualdades no acesso à saúde suplementar. A distribuição desigual dos planos reflete desafios estruturais que podem comprometer a inclusão de diferentes grupos populacionais. A pandemia evidenciou essas discrepâncias e destacou a importância da acessibilidade aos serviços de saúde. Assim, identificar padrões e disparidades pode revelar lacunas que demandam ações estratégicas.

Esse estudo justifica-se pelo seu impacto social e econômico. Do ponto de vista social, entender a distribuição dos beneficiários e os fatores que influenciam o acesso aos planos de saúde pode subsidiar a formulação de políticas públicas mais eficazes, promovendo inclusão e equidade. No aspecto econômico, o setor de saúde suplementar movimenta bilhões de reais anualmente, sendo essencial compreender os padrões de crescimento e sustentabilidade financeira. Além disso, a aplicação de técnicas de Ciência de Dados e análise de séries temporais permite gerar insights relevantes para tomadores de decisão, reguladores e operadoras de planos de saúde.

O objetivo geral deste estudo é analisar a evolução do número de beneficiários de planos de saúde no Brasil entre 2019 e 2024, identificando padrões, tendências e fatores que influenciam o crescimento ou estagnação do setor, com foco em faixas etárias, tipos de contratação e distribuição geográfica por Unidade da Federação.

Os objetivos específicos incluem:

Investigar o perfil demográfico dos beneficiários, considerando faixa etária, localização e tipo de contratação;

Aplicar técnicas de análise de séries temporais para avaliar a evolução dos planos de saúde ao longo do período analisado;

Identificar possíveis correlações entre o crescimento de beneficiários e fatores socioeconômicos ou eventos relevantes, como a pandemia de Covid-19;

Gerar visualizações e relatórios que auxiliem operadoras e órgãos reguladores na tomada de decisão.

Para a realização deste estudo, serão utilizadas bases de dados fornecidas pela ANS e outros órgãos do setor. As bases incluem:

Número de Beneficiários em Planos de Assistência Médica, segundo as Faixas Etárias: Dados que descrevem a quantidade de usuários por faixa etária, permitindo análises do perfil demográfico.

Número de Beneficiários com Planos de Assistência Médica por Tipo de Contratação: Informam quantos beneficiários possuem planos individuais/familiares, coletivos empresariais ou coletivos por adesão.

Número de Beneficiários por UF: Indicam a distribuição geográfica dos beneficiários ao longo do período analisado, permitindo avaliar variações regionais.

As bases de dados contêm colunas como “Data de Referência”, “Faixa Etária”, “Tipo de Contratação” e “UF”, permitindo análises detalhadas. O volume de informações varia conforme a granularidade (mensal ou anual), e algumas lacunas podem ser identificadas e tratadas na etapa de limpeza de dados.

Com esta abordagem, espera-se fornecer uma visão abrangente sobre a dinâmica do setor de planos de saúde no Brasil, gerando subsídios para a formulação de estratégias mais eficazes e inclusivas.

Em problemas semelhantes, há estudos que empregam regressão linear para relacionar crescimento de beneficiários a variáveis macroeconômicas (MARTINS et al., 2020), bem como modelos de série temporal (ARIMA, SARIMA) para prever flutuações no número de usuários (SANTOS; PEREIRA, 2019). Uma vantagem dos modelos de série temporal é a capacidade de capturar padrões cíclicos e sazonais, porém podem apresentar limitações ao lidar com choques exógenos repentinos, como crises sanitárias ou mudanças regulatórias (ANS, 2023). Por outro lado, técnicas de aprendizado de máquina conseguem integrar múltiplas variáveis de forma mais flexível, mas podem demandar grande volume de dados e maior complexidade computacional (RIBEIRO; ALMEIDA, 2022).

# **Pipeline da Solução**

2.1. Pipeline de Análise

A solução proposta visa integrar e analisar os dados disponibilizados (número de beneficiários por faixa etária, UF e tipo de contratação) por meio de técnicas de série temporal. O fluxo de trabalho segue as etapas abaixo:

Coleta e Integração dos Dados

Reunir as planilhas fornecidas (beneficiários por UF, faixa etária, tipo de contratação) em um único repositório.
Padronizar os formatos de data, remover duplicações e corrigir inconsistências.
Limpeza e Preparação

Tratar valores ausentes ou inválidos (ex.: substituição por média, mediana ou exclusão, dependendo do contexto).
Criar colunas derivadas (ex.: “Ano/Mês” consolidado) para facilitar a análise temporal.
Análise Exploratória

Visualizar a evolução do número de beneficiários em gráficos de linha, comparando diferentes categorias (UFs, faixas etárias, tipos de contratação).
Calcular estatísticas descritivas (média, mediana, desvios) para cada subgrupo.
Modelagem de Série Temporal

Escolher modelos de previsão adequados (ARIMA, SARIMA, Holt-Winters, Prophet etc.).
Ajustar hiperparâmetros e comparar o desempenho de cada modelo (usando métricas como MAPE, RMSE).
Validação e Interpretação

Avaliar a acurácia dos modelos em dados de teste.
Interpretar tendências (ex.: crescimento acelerado em determinados estados ou faixas etárias).
Geração de Relatórios e Insights

Criar gráficos, tabelas e painéis (dashboards) para sintetizar os achados.
Apresentar recomendações para operadoras, gestores públicos e reguladores.

2.2. Ferramentas e Técnicas

Linguagem de Programação: Python (bibliotecas como pandas, numpy, matplotlib, seaborn, statsmodels ou Prophet).
Processo de Ciência de Dados: CRISP-DM ou KDD, para guiar o ciclo de descoberta de conhecimento.
Visualizações: Gráficos de linha para séries temporais, histogramas para distribuições, mapas coropléticos para distribuição geográfica.
